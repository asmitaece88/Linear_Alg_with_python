Understanding the concept of Matrices 

Consider a Scenario,

Lets say Claire has 5 note books and 2 pens.

We have learnt already in previous section of Vectors, We can represent this data in the form of Vector as a List.




Now, we have few more data,Alex has 15 note books and 10 pens.
Phil has 2 note books and 12 pens.
If we arrange these data in tabular form,It would look like this,


This data can be expressed as,Data enclosed within square brackets like this.



In the first arrangement, the first column represents the number of notebooks owned by Claire, Alex and Phil Respectively, and second column represents the number of pens owned by them respectively. 

Now, in the second arrangement ,The columns represent the number of notebooks and pens owned by Claire, Alex and Phil respectively.

This arrangement of data in horizontal rows and Vertical columns is called as Matrix.

Now, to define the Matrices formally,

A  Matrix is a two-dimensional array of scalars with one or more columns and one or more rows.

Now, the next basic concept is Order of Matrices.
A matrix having m rows and n columns is called a matrix of order m × n or simply m × n matrix

By looking at the above matrices which we have defined previously, We can say that, these Matrices are of order 3 by 2 and 2 by 3 matrices respectively.


This is the representation of m cross n matrix.

That is M rows and N columns.

This representation of Matrix by number of rows and number of columns is known as Dimension of matrix.
